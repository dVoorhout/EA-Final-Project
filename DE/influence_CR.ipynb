```python
import matplotlib.pyplot as plt
import numpy as np

from math import sqrt
from time import sleep, time
from random import seed


from DE import DifferentialEvolution
from Packomania import Packomania
from utils import *
```

```python
seed(0)

num_points = 3
bounds = [[0, 1] for i in range(num_points*2)]
pk = Packomania(num_points)
optimum = pk.points_min_dist

differential_weight = 0.5
crossover_prob_list = [0.1, 0.3, 0.5, 0.8, 0.9,1.0]        
popsize = 10*num_points*2 # *2 since dimension is num_points*2

max_evaluations = 5e6
optimum_accuracy = 1e-6
results = []


start_time = time()
for crossover_prob in crossover_prob_list:
    print(f"Crossover_prob: {crossover_prob}")
    de = DifferentialEvolution(
                       scattering_points, 
                       bounds, 
                       popsize, 
                       differential_weight, 
                       crossover_prob
    )
    de.set_optimum(optimum, optimum_accuracy)
    failed_flag, r = find_reliable(algo=de, stop_on_fail=False, max_evaluations=max_evaluations, verbose=False)
    results.append(r)
print(f"TOTAL RUNTIME: {time()-start_time}s")
```

```python
save_reliable_results(f"Params/num_points_{num_points}_{optimum_accuracy}_CR2.csv", crossover_prob_list, results)
```
