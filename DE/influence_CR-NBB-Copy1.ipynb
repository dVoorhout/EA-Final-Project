```python
import matplotlib.pyplot as plt
import numpy as np

from math import sqrt
from time import sleep, time
from random import seed


from DE import DifferentialEvolution
from Packomania import Packomania
from utils import *
from benchmark import benchmark, get_popsize
```

```python
seed(0)

num_points = 7
bounds = get_stricter_bounds(num_points)
pk = Packomania(num_points)
optimum = pk.points_min_dist

crossover_prob_list = [0.8, 0.9, 0.99]  
get_popsize_func_A3 = lambda dim: get_popsize(dim=dim, m=30)

results_A3 = []


for crossover_prob in crossover_prob_list:
    print(f"Crossover {crossover_prob}")
    r_A3 = benchmark(
        differential_weight=0.1,
        crossover_prob=crossover_prob,
        max_evaluations=5e6,
        get_popsize_func=get_popsize_func_A3,
        get_init_bounds_func = get_stricter_bounds,
        num_points_list=[num_points],
        min_pop_variance=1e-25,
        crossover='best_points3',
        optimum_accuracy=1e-6
    )
    results_A3.append(r_A3[0])
```
