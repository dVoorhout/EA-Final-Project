```python
import matplotlib.pyplot as plt
import numpy as np

from math import sqrt
from time import sleep, time
from random import seed


from DE import DifferentialEvolution
from Packomania import Packomania
from utils import *
```

```python
seed(0)

num_points = 3
bounds = [[0, 1] for i in range(num_points*2)]
pk = Packomania(num_points)
optimum = pk.points_min_dist

differential_weight_list = [0.1, 0.3, 0.5, 0.8, 1.0]         
crossover_prob = 0.3
popsize = 10*num_points*2 # *2 since dimension is num_points*2

max_evaluations = 5e6
optimum_accuracy = 1e-6
results = []


start_time = time()
for differential_weight in differential_weight_list:
    print(f"Differential_weight: {differential_weight}")
    de = DifferentialEvolution(
                       scattering_points, 
                       bounds, 
                       popsize, 
                       differential_weight, 
                       crossover_prob
    )
    de.set_optimum(optimum, optimum_accuracy)
    failed_flag, r = find_reliable(algo=de, stop_on_fail=False, max_evaluations=max_evaluations, verbose=False)
    results.append(r)
print(f"TOTAL RUNTIME: {time()-start_time}s")
```

```python
save_reliable_results(f"Params/num_points_{num_points}_{optimum_accuracy}_F.csv", differential_weight_list, results)
```
